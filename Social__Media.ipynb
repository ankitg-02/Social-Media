```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing data from csv file
d=pd.read_csv(r'/content/Time-Wasters on Social Media.csv')
data=pd.DataFrame(d)
data.head()
```

```python
#data types of columns of the dataset
data.dtypes
```

```python
#information about dataset
data.info()
```

```python
#checking the null values in the dataset
data.isnull().sum()
```

```python
for i in data.columns:
  print(f"{i}: {data[i].unique()}\n")
```

```python
plt.figure(figsize=(10,10))
sns.displot(data['Time Spent On Video'])
plt.title('Distribution of Time Spent on Video')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.distplot(data['Addiction Level'])
plt.title('Distribution of Addiction Level on Video')
plt.show()
```

```python
data['Debt'].value_counts().plot(kind='bar')
plt.title('Distribution of Debt')
plt.show()
```

```python
data['Video Category'].value_counts().plot(kind='bar')
plt.title('Distribution of users a/c to Video Category')
plt.show()
```

```python
data['Location'].value_counts().plot(kind='bar')
plt.title('Distribution of Location')
plt.show()
```

```python
data['Profession'].value_counts().plot(kind='pie',autopct="%1.1f%%")
plt.title('Distribution of Profession')
plt.show()
```

```python
data['DeviceType'].value_counts().plot(kind='pie',autopct="%1.1f%%")
plt.title('Distribution of DeviceType users')
plt.show()
```
